# Importing all the basic libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

# Starting the Exploratory Data Analysis (EDA)

---

```python
data = pd.read_csv('/content/kag_risk_factors_cervical_cancer.xls')
data = data.replace("?",np.nan) # all the null values already have been replaced by '?'
print(data.isnull().sum())
data
```

```python
data.drop(['STDs: Time since first diagnosis','STDs: Time since last diagnosis'], axis = 1,inplace = True) # These columns have no relevance in our prediction. So removed them.
data = data.drop_duplicates() #dropping the duplicates
data.dtypes
data
```

# Data Pre-processing

---



### Here we convert tha datatype of each Series to numeric (either float64 or int64)

```python
data_num = data.select_dtypes(include = "number")
data_obj = data.select_dtypes(include  = "object")
data_obj = data_obj.apply(pd.to_numeric)
data = pd.concat([data_num,data_obj], axis = 1, join="inner")

data.dtypes
```

### Now we replace all the missing values (NaN values) with the help of Scikitlearn Imputer.

### The imputer replaces all the np.nan values with the mean/median of the respective column

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
data_new = pd.DataFrame(imputer.fit_transform(data))
data_new.columns = data.columns
data_new
```

### checking if PCA can be applied to the above data set

### PCA can only be applied if most of the correlation coefficients are > 0.3

```python

corrmat = data_new.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))

#plot a heat map for a better visualisation of the correlations co-efficients
g=sns.heatmap(data_new[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

### Here we scale down different features to the same scale. This would not let any feature with a larger range to
### dominate over any other feature which has a smaller range.

```python
from sklearn.preprocessing import RobustScaler

y = data_new['Biopsy']
X = data_new.drop(['Biopsy'], axis = 1)

cols = X.columns

my_scaler = RobustScaler()

X_train = my_scaler.fit_transform(X)

X_train = pd.DataFrame(X)

X.head()
```

### Now we convert the imbalanced dataset to a balanced one using Synthetic Minority Over-sampling Technique (SMOTE)

> ## a brief description about the SMOTE Technique:


### SMOTE is also a type of Over-sampling Technique but the catch is that we create synthetic data in SMOTE instead of trivially repeating them.
### We use Linear Algebra to find a vector between the selected minority datapoint and the nearest centroid (or the respective cluster centroid.)
### We then multiply that vector by a number between 0 and 1 then add this is to the original minority data-point. Thus creating a synthetic data point.

```python

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_new, y_new = sm.fit_resample(X, y)
X_new
```

# Selecting 10 best Features
---

### Now we start the process of best feature selection using 'feature_selection' method of the sklearn library
### Also we can't use PCA to because of the aforementioned rationale.


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split

X_train,X_test,y_train, y_test = train_test_split(X_new,y_new, test_size = 0.33, random_state = 1234)

bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X_train, y_train)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

# concat two dataframes for better visualization

featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

### Above, we have got top 10 best features
### Let's extract these features to a new dataset. This new dataset will be used to train our model.

```python

new_features = featureScores.nlargest(10,'Score')['Specs']
new_data2 = pd.DataFrame(X_new[new_features])

print(new_data2)
```

# The Training of different models

---

### Now I will start training different models.
### The metric that I will be using to evaluate the models is the F1 score. The model with the best F1 score will be selected.

# ***The Naive Bayes Classifier***

```python
X_new = new_data2

X_train,X_test,y_train, y_test = train_test_split(X_new,y_new, test_size = 0.33, random_state = 1234)

# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB

# instantiate the model
gnb = GaussianNB()


# fit the model
gnb.fit(X_train, y_train)
y_pred_Naive_bayes = gnb.predict(X_test)

from sklearn.metrics import f1_score
print("the F1 score of the Naive Bayes model is: ", f1_score(y_test,y_pred_Naive_bayes))
```

# ***The Support Vector Machine (SVM) Classifier***

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', random_state = 1)
classifier.fit(X_train,y_train)

y_pred_SVM = classifier.predict(X_test)
print("The F1 score of the SVM model is:" ,f1_score(y_test,y_pred_SVM))
```

# ***The Random Forest Classifier***

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 1234)
model.fit(X_train,y_train)

y_pred_RFC = model.predict(X_test)

print("The F1 score of the Random Forest Classifier is :", f1_score(y_test,y_pred_RFC))
```

# ***The decision Tree Classifier***

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy' , splitter = 'best', random_state = 1234)

model.fit(X_train,y_train)

y_pred_DTC  = model.predict(X_test)

print("The F1 score of the Decision Tree Classifier is:", f1_score(y_test,y_pred_DTC))
```

# ***The Logistic Regrssion model***

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_test,y_test)

y_pred_LR = model.predict(X_test)
print("The F1 score of the Logistic model is:", f1_score(y_test,y_pred_LR))
```

# ***The K-nearest Nieghbor Classifier***

```python
from sklearn.neighbors import KNeighborsClassifier

model  = KNeighborsClassifier()

model.fit(X_train,y_train)
y_pred_KNN = model.predict(X_test)
print("The F1 score of the KNN model is:", f1_score(y_test,y_pred_KNN))
```
